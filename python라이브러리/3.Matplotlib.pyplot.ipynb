```python
import pandas as pd
```

```python
data = pd.read_csv('Traffic_Accident_2017.csv', encoding ='euc-kr')
data.head()
```

```python
#data['요일']
#각각의 요일이 몇번씩 있는지 세기
temp = data['요일'].value_counts()
#데이터를 요일 순으로 출력
y= temp[['월','화','수','목','금','토','일']]
```

```python
x = y.index

# x축 이름: 요일
# y축 이름: 사고 건수
# 전체이름: 요일별 교통사고 건수
plt.bar(x,y)
plt.xlabel('요일',position=(0.3, 0.55))
plt.ylabel('사고건수',position=(0.5, 0.55)).set_rotation(0)  #position=(x축은 고정,y)
plt.title('교통사고건수') 

plt.ylim(500,620)
plt.show()
```

```python
# data['사고유형_대분류']
# 차대차
car_to_car = data[data['사고유형_대분류'] == '차대차']
```

```python
#car_to_car
#사상자수, 발생지시도 출력
#groupby('발생지시도') :같은 발생지시도 값을 가지고 있는 데이터끼리 묶깊
car_to_car_place = car_to_car[['사상자수','발생지시도']].groupby('발생지시도').sum()
```

```python
x= car_to_car_place.index
y= car_to_car_place.values
```

```python
plt.figure(figsize=(15,18)) #그래프 크기 조절/ 가장 먼저 적어야함
plt.plot(x,y)
plt.show()
```

```python
# bar 차트 그리기
y[0][0]
```

```python
# 배열 안에 담기 배열값을 하나씩 꺼내주기
# for문으로 접근하기
y_bar = []
for i in range(len(y)):
    y_bar.append(y[i][0])
```

### for i in range(len(y)) 로 적어주셔야하는데
### for i range(len(y))로 적으셔서 그런거같아요 !

```python
y_bar
```

```python
plt.figure(figsize =(15,8))
plt.bar(x,y_bar)
plt.xlabel('지역')
plt.ylabel('사상자수').set_rotation(0)
plt.title('차 vs 차 교통사고의 사상자수')
plt.show()
```

### 교통사고가 가장 많이 발생하는 시간대 알아보기
- 발생년월일시

```python
time = data['발생년월일시']%100    #나머지(시간 데이터만 담겨 있다)
```

```python
time
```

```python
# 구간    #이름
# 0~2     0~2
# 3~5
# 6~8
# 9~11
# 12~14
# 15~17
# 18~20
# 21~23
# 카테고리 데이터 만들기
```

```python
bins = [-1, 2, 5, 8, 11, 14, 17, 20, 23]   #범위(ex. -1초과 2이하, 2초과 5이하..)
labels = ['0~2','3~5','6~8','9~11','12~14','15~17','18~20','21~23']

result = pd.cut(time, bins = bins, labels = labels)
result

# 시간별로 몇 건의 사고가 발생했는지 세보기 -> 시간 순으로 정렬
print(result.value_counts())
print('-------------------------------')
print(result.value_counts().sort_index())
```

```python
rs = result.value_counts().sort_index()
plt.figure(figsize= (5,5))
# labels: 각 데이터에 이름 추가
# autopct: 각 데이터 몇 %차지하고 있는지(-> 소숫점 2째자리 + 표시)   # series: value+index
plt.pie(rs,labels = labels, autopct = '%.2f%%')
plt.show()
```
