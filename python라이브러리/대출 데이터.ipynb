## 대출 기간에 따른 대출 총액
- 대출기간: term
- 36개월, 60개월

```python
df['term'].value_counts()
```

```python
# 대출기간당 총액을 담을 딕셔너리 생성
term_to_loan = {}
# 반복을 돌릴 값 생성
uniq_terms = df['term'].unique()
# for문을 통해 값 생성
for term in uniq_terms:
    term_to_loan_sum = df.loc[df['term']==term , 'loan_amnt'].sum()
    term_to_loan[term] = term_to_loan_sum  #딕셔너리에 데이터 담기
# 결과값 Series에 담아주기
term_to_loan_series = pd.Series(term_to_loan)
# 결과값 DataFrame에 담아주기
result = pd.DataFrame(term_to_loan_series,columns = ['loan_amnt'])
result.index.name = 'term'
result
```

# 대출 이자율을 상,중,하로 구분하고 구분한 값을 새로운 컬럼을 만들어서 대입, -> 각 요소의 대출액 평균 구하기
- 대출이자율 : int_rate
- apply
- cut

```python
df['int_rate'].describe()
```

```python
#apply함수에 적용할 함수 만들기
def make_standard(apt):
    if apt['int_rate'] < 9:
        return 'good'
    elif (apt['int_rate'] >=9) & (apt['int_rate'] < 12):
        return 'soso'
    else :
        return 'bad'
```

```python
df['int_rate_grade1'] = df.apply(make_standard, axis=1)
```

```python
df[['loan_amnt','int_rate_grade1']].groupby('int_rate_grade1').mean()
```

```python
bins = [5,9,12,29]
label = ['good','soso','bad']
df['int_rate_grade'] = pd.cut(df['int_rate'],bins = bins, labels = label)

grade_term = df['int_rate_grade'].unique()

grade_mean = {}
for grade in grade_term:
    grade_by_mean = df.loc[df['int_rate_grade'] == grade,'loan_amnt'].mean()
    grade_mean[grade] = grade_by_mean

grade_mean
```
